# Функции

```python
def correct_colname(df):
    df.columns = df.columns.str.lower().str.replace(" ", "_")


def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 16)
            .set_title('Доля пропусков в столбцах датафрейма (%)' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('Пропусков нет')


def check_and_remove_duplicates(df, subset=None, keep='first', inplace=True):
    duplicates_count = df.duplicated(subset=subset).sum()
    
    if duplicates_count > 0:
        if inplace:
            df.drop_duplicates(subset=subset, keep=keep, inplace=True)
            result_df = df
        else:
            result_df = df.drop_duplicates(subset=subset, keep=keep)
    else:
        result_df = df
    return result_df


def check_emissions(df):
    numeric_cols = df.select_dtypes(include=['number']).columns
    print(df[numeric_cols].describe())


def remove_outliers_iqr(df, column, iqr_multiplier=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - iqr_multiplier * IQR
    upper_bound = Q3 + iqr_multiplier * IQR
    df_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)].copy()
    
    return df_clean


def remove_single_max(df, column):
    max_value = df[column].max()
    df_clean = df[df[column] != max_value].copy()
    return df_clean


def unique_elements(df):
    for i in df.columns:
        print(df[i].unique()[:10])


def build_histogram(df, col_1, col_2):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(data=df, x=col_1, kde=True, ax=ax1)
    ax1.set_title(f'Распределение {col_1}')
    sns.histplot(data=df, x=col_2, kde=True, ax=ax2)
    ax2.set_title(f'Распределение {col_2}')
    plt.tight_layout()
    plt.show()


def build_scatterplot(df, col1, col2):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x=col1, y=col2, alpha=0.6)
    sns.regplot(data=df, x=col1, y=col2, scatter=False, 
                line_kws={'color': 'red', 'linewidth': 2})
    plt.title(f'{col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.grid(True, alpha=0.3)
    plt.show()
    
def calculation_correlation(df, col1, col2):
    _, p1 = stats.shapiro(df[col1])
    _, p2 = stats.shapiro(df[col2])
    
    if p1 > 0.05 and p2 > 0.05:
        method = 'pearson'
        corr, p_value = stats.pearsonr(df[[col1, col2]].dropna())
    else:
        method = 'spearman'
        corr, p_value = stats.spearmanr(df[col1], df[col2])
    
    print(f"{col1} vs {col2}")
    print(f"Коэффициент корреляции: {corr:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value > 0.05:
        if abs(corr) < 0.1:
            print("Корреляция близка к нулю и статистически не значима")
            print("Переменные действительно не связаны")
        else:
            print("Корреляция есть, но статистически не значима")
            print("Возможно: малая выборка или случайная флуктуация")
    else:
        print("Корреляция статистически значима")
        if abs(corr) < 0.3:
            print(" Слабая связь")
        elif abs(corr) < 0.7:
            print("Умеренная связь")
        else:
            print("Сильная связь")
    
    return corr, p_value


def analyze_pair(df, col1, col2):
    print(f"АНАЛИЗ: {col1} и {col2}")
    
    build_histogram(df, col1, col2)
    
    build_scatterplot(df, col1, col2)
    
    calculation_correlation(df, col1, col2)


def cycle(pair):
    for col_1, col_2 in pair:
        
        analyze_pair(df, col_1, col_2)
```

# ВЫВОДЫ ПО ВСЕЙ РАБОТЕ

Маркетинг устарел — навязчивые коммуникации отталкивают, а не привлекают.
VIP-клиенты в зоне риска — самые прибыльные покупатели с наибольшей вероятностью ухода.
Отсутствует персонализация — единый подход ко всем клиентам неэффективен.
Акционная зависимость подрывает лояльность и долгосрочную прибыльность.

Необходимо полностью пересмотреть подход к удержанию клиентов, сместив фокус с массового маркетинга на персонализированный для ценных сегментов. Также необходимо обратить внимание на популярные товары у VIP-клиентов, а именно косметика и аксессуары, товары для детей и домашний текстить, стоит сделать акцент на эти товары, возможно ввезти новые уникальные виды, создать индивидуальные предложения для VTP. 

Возможная послежоватлеьность действий:
Прекратить маркетинговые манипуляции(рассылки, спам, постоянных скидок и др.)
Выделить самых ценных клиентов,те кто принесут 80-90% всей прибыли и сконцентрироваться на них.
В идеале назначить им помощников, но не навязчиво, которые бы консультировали и др.
Делать персональные предлодения, продавать уникальный товар, товар с фашками, который бы мог заинтересовать VIP.
Бросить практически все силы на них, оставшиеся бросить на остальных, чтобы извлечь максимум по возможности и др. Двигаться в этом направлении пока что.
